## Modele generujące iglice

Iglica, jak podaje Enyklopedia PWN, to ,,podłoże nerwowego impulsu; polega na dużym i gwałtownym obniżeniu potencjału błonowego wskutek otwarcia kanałów sodowych i wniknięciu jonów sodu do komórki''. W 1907r., francuski neurobiolog, Louis Édouard Lapicque zaproponował model neuronu (integrate-and-fire). Jest to najprostszy model, który generuje iglicę. Jest on opisany równaniem:
<br>
$$C\frac{dV(t)}{dt} = I(t),$$
<br>
gdzie V(t) - potencjal błonowy, a I(t) - napięcie wejściowe (suma napięcia pobujdzającego i hamującego).
Podobny do powyższego, jednak uzupełniony o dyfundowanie jonów przez membranę, jest model nazywany ,,leaky integrate and fire'', który opisujemy równaniem:
<br>
$$C\frac{du(t)}{dt} = I(t) - \frac{u(t) - u_{rest}}{R}. $$
<br>
Powyższe równanie mnożymy przez R (opór membrany) oraz wprowadzamy $\tau_m = RC$ i otrzymujemy równanie (1):
<br>
$$\tau_m \frac{du(t)}{dt} = -[u(t) - u_{rest}] + RI(t). $$
<br>
Pezy założeniu, że $t = 0$ potencjał membrany przyjmuje wartość $u(t_0) = u_{rest} + \Delta u$. Zatem otrzymujemy równanie postaci:
<br>
$$u(t) - u_{rest} = \Delta u \exp\left(- \frac{t - t_0}{\tau_m}\right),$$ dla $t > 0$.
<br>
Z równania (1) opisującego dynamikę membrany biernej, poprzez całkowanie z warunkiem początkowym $u(0) = u_{rest} $ otrzymamy trajektorię potencjału błonowego, dla $0 < t < \Delta$:
<br>
$$u(t) = u_{rest} + RI_0\left[1 - \exp\left(\frac{-t}{\tau_m}\right)\right].$$
<br>


## Model Hodgkina-Huxleya

Najczęściej używanym modelem służącym do opisu transmisji impulsów nerwowych jest model Hodgkina-Huxleya, za którego stworzenie naukowcy otrzymali nagrodę Nobla w dziedzinie medycyny w 1963 roku. Model ten, stworzony w 1952 roku, naśladuje rzeczywiste zachowania neuronu. W modelach matematycznych możemy uzyskać większą dokładność przez podział włókien nerwowych komórek na mniejsze, które później opisujemy odpowiednimi równaniami, porównując je do obwodów elektrycznych. Obwody te jednak należy stworzyć przez taki podział, aby analogiczne obwody RC posiadały w przybliżeniu stały potencjał oraz były zgodne z topologią dendrytów. Następujący rysunek pokazuje przykład takiego obwodu.
 

Można zauważyć, że prąd ładujący pojemność błonową to $C_{m} \frac{dV_{m}}{dt} + I_{ion} = I_{ext}$, przy czym $I_{ion} = G_{Na}(V_{m} - E_{Na}) + G_{K}(V_{m} - E_{K}) + G_{L}(V_{m} - E_{L})$. Prąd jonowy $I_{k}$ zależy od przewodności oraz potencjału równowagowego w poniższy sposób: $I_{k} = G_{k}(V_{m} - E_{k}).$ Gdzie $C_m$ to pojemność membrany, $G_{Na}$ i $G_K$ to przewodności bramek odpowiednio sodowych i potasowych, $G_{L}$ to przewodność upływu, $E_{Na}$, $E_K$ oraz $E_L$ to potencjały równowagowe odpowiednich kanałów, a $V_{m}$ to wartość potencjału błony komórkowej.
Aby mógł zajść przepływ wszystkie bramki muszą być w stanie ,,przyzwalającym''. Prawdopodobieństwo otwarcia bramki danego typu opisuje wzór:
$\frac{dp_{i}}{dt} = \alpha_{i}(V)(1-p_{i})-\beta_{i}(V)p_{i},$
przy czym parametry $\alpha$ oraz $\beta$ to stałe zależne od napięcia.
W modelu zakładamy, że kanał sodowy składa się z trzech bramek $m$ i jednej $h$, a potasowy z czterech bramek $n$, zatem:
$$G_{Na} = \overline{g}_{Na} p_{m}^3 p_{h} \equiv \overline{g}_{Na} m^3 h,$$
<br>
$$G_{K} = \overline{g}_{K} p_{n}^4 \equiv \overline{g}_{K} n^4.$$
<br>
Po podstawieniu przewodności $G$ otrzymujemy równania Hodgkina-Huxleya:
<br>
$$C_{m} \frac{dV_{m}}{dt} + I_{ion} = I_{ext},$$
<br>
$$I_{ion} = \overline{g}_{Na} m^3 h(V_{m} - E_{Na}) + \overline{g}_{K} n^4(V_{m} - E_{K}) + \overline{g}_{L}(V_{m} - E_{L}),$$
<br>
oraz mamy trzy typy bramek, dla których prawdopodobieństwo otwarcia zależy od napięcia w następujący sposób:
<br>
$$\frac{dm}{dt} = \alpha_{m}(V)(1-m)-\beta_{m}(V)m,$$
<br>
$$\frac{dh}{dt} = \alpha_{h}(V)(1-h)-\beta_{h}(V)h,$$
<br>
$$\frac{dn}{dt} = \alpha_{n}(V)(1-n)-\beta_{n}(V)n.$$
<br>
Mamy również doświadczalnie wyznaczone funkcje wykładnicze $\alpha$ i $\beta$:
<br>
$$\alpha_n(V_m) = \frac{0,01 (V_m + 55)}{[1-\exp(-V_m + 55) \frac{1}{10}]},$$
<br>
$$\alpha_m(V_m) = \frac{0,01 (V_m + 40)}{[1-\exp(-V_m + 40) \frac{1}{10}]},$$
<br>
$$\alpha_h(V_m) = 0,07 \exp{[-(V_m + 65) \frac{1}{20}]},$$
<br>
$$\beta_n(V_m) = 0,125 \exp{[-(V_m + 65) \frac{1}{80}]},$$
<br>
$$\beta_m(V_m) = 4 \exp{[-(V_m + 65) \frac{1}{18}]},$$
<br>
$$\beta_h(V_m) = \frac{1}{[1+\exp(-V_m + 35) \frac{1}{10}]},$$
<br>
oraz stałe:
$G_{Na} = 120 mS/cm^2$, $G_K = 36 mS/cm^2$, $G_L = 0,3 mS/cm^2$, $E_{Na} = 50 mV$, $E_K = -77 mV$, $E_L = -54 mV$, $C_m = 1 \mu F/cm^2$.

### Rozwiązywanie równań motodą Eulera
Wychodzimy od równania:
<br>
$$C_{m} \frac{dV_{m}}{dt} = - G_{Na}(V_{m} - E_{Na}) - G_{K}(V_{m} - E_{K}) - G_{L}(V_{m} - V_{L}).$$
<br>
Dzieląc obustronnie przez C:
<br>
$$\frac{dV_{m}}{dt} = \left(- G_{Na}(V_{m} - E_{Na}) - G_{K}(V_{m} - E_{K}) - G_{L}(V_{m} - V_{L})\right) \frac{1}{C_{m}}.$$
<br>
Definiujemy funkcję f postaci:
<br>
$$f(V_{m}) = \left(- G_{Na}(V_{m} - E_{Na}) - G_{K}(V_{m} - E_{K}) - G_{L}(V_{m} - V_{L})\right) \frac{1}{C_{m}},$$
<br>
$$f(V_{m}) = V_{m}\left(-\frac{G_{Na}}{C_{m}}-\frac{G_{K}}{C_{m}} - \frac{G_{L}}{C_{m}}\right) + \frac{G_{Na}}{C_{m}} E_{Na} + \frac{G_{K}}{C_{m}} E_{K} + \frac{G_{L}}{C_{m}} V_{L}.$$
<br>
Zatem:
<br>
$$\frac{dV_{m}}{dt} = f(V_{m})$$
<br>
Całkujemy obustronnie po czasie przyjmując granice całkowania $[t, t + \Delta t]$:
<br>
$$\int_{t}^{t + \Delta t} \frac{\partial V_{m}}{\partial \tau} d \tau = \int_{V_{m}(t)}^{V_{m}(t + \Delta t)} d \tau =  V_{m}(t + \Delta t) - V_{m}(t).$$
<br>
Zauważmy, że korzystając z przybliżania całki metodą prostokątów, możemy zapisać, że dla bardzo małych wartości $\Delta t$:
<br>
$$\int_{t}^{t + \Delta t} f[V_{m}(\tau)] d \tau \approx \Delta t \cdot f[V_{m}(t)].$$
<br>
Zatem:
<br>
$$V_{m}(t + \Delta t) + V_{m}(t) = \Delta t \cdot f[V_{m}(t)].$$
<br>
Odejmując obustronnie $V_{m}(t)$ oraz podstawiając funkcję $f[V_{m}(t)]$ otrzymujemy:
<br>
$$V_{m}(t + \Delta t) = \Delta t \left(V_{m}(t)\left(1 - \frac{G_{Na} + G_{K} + G_{L}}{C_{m}}\right) + \frac{G_{Na}}{C_{m}} E_{Na} + \frac{G_{K}}{C_{m}} E_{K} + \frac{G_{L}}{C_{m}} V_{L}\right).$$
<br>
W ten sam sposób można rozwiązać również równania na n, m oraz h. Dla uproszczenia opisu zauważmy, że równania te różnią się tylko postaciami odpowiednich funkcji $\alpha$ i $\beta$. Rozwiążemy zatem równanie na x:
<br>
$$\frac{dx}{dt} = \alpha_x(V_{x})(1-x) - \beta_x(V_{m})x.$$
<br>
Aby opis był wygodniejszy, oznaczmy lewą stronę jako $g(x)$
<br>
$$\frac{dx}{dt} = \alpha_x + x\left(-\alpha_x-\beta_x\right) = g(x).$$
<br>
Obie strony całkujemy tak, jak wcześniejsze równanie oraz korzystamy z przybliżania metodą prostokątów i otrzymujemy:
<br>
$$x(t + \Delta t) - x(t) = \Delta t \cdot g[x(t)].$$
<br>
Po podstawieniu funkcji $g$ oraz przeniesieniu x(t) na drugą stronę otrzymujemy:
$$x(t + \Delta t) = \alpha_x + x\left(1 - \alpha_x -\beta_x\right).$$

```python
import matplotlib.pyplot as plt
import math
```

```python
# Implementacja stałych

g_k = 36.0
g_na = 120.0
g_l = 0.3
c_m = 1.0
e_na = 50.0
e_k = -77.0
v_l = -54.387
v_init = -70.0
tau = (g_na + g_k + g_l)/c_m
delta_t = 0.006
t = [0]
for i in range(0, 10000):
    t.append(t[i]+0.01)
```

```python
def voltage(v):
    return 0.006*(v*(1-tau) + (g_na/c_m)*e_na + (g_k/c_m)*e_k + (g_l/c_m)*v_l)

def a_n(v):
    return (0.01*(v+55))/(1-(math.exp(-(v+55)))*0.1)

def a_m(v):
    return (0.01 * (v + 40)) / (1 - (math.exp(-(v + 40)))*0.1)

def a_h(v):
    return 0.07 * math.exp(-(v + 65) * 0.05)

def b_n(v):
    return 0.125 * math.exp(-(v + 65) * 0.0125)

def b_m(v):
    return 4 * math.exp(-(v + 65) * 1/18)

def b_h(v):
    return 1 / (1 + (math.exp(-v + 35)) * 0.1)

def n_gate(n, v):
    return a_n(v) + n*(1 - a_n(v) - b_n(v))

def m_gate(m, v):
    return a_m(v) + m * (1 - a_m(v) - b_m(v))

def h_gate(h, v):
    return a_h(v) + h * (1 - a_h(v) - b_h(v))
```

```python
v_val = [v_init]
n_val = [0.5]
m_val = [0.5]
h_val = [0.5]
v_x = [v_init]
for i in range(0, 10000):
    v_x.append(t[i]+0.1)
def calculate():
    for i in range(len(t) - 1):
        v = voltage(v_val[i])
        v_val.append(v)
        n = n_gate(n_val[i], v)
        n_val.append(n)
        m = m_gate(m_val[i], v)
        m_val.append(m)
        h = h_gate(h_val[i], v)
        h_val.append(h)
    return v_val, n_val, m_val, h_val

def main():
    v_val, n_val, m_val, h_val = calculate()
    plt.plot(t, v_val)
    plt.xlim(0, 0.5)
    plt.title('Wykres V(t)')
    plt.xlabel('t(ms)')
    plt.ylabel('V(mV)')
    plt.show()
    plt.plot(v_x, n_val, label='n')
    plt.plot(v_x, m_val, label='m')
    plt.plot(v_x, h_val, label='h')
    plt.xlim(0, 1)
    plt.title('Wartości bramkowania')
    plt.xlabel('V(mV)')
    plt.legend()
    plt.show()
```

```python
if __name__=="__main__":
    main()
```

## Redukcja równań Hodgkina-Huxleya
Równania na m, n lub h możemy przepisać do postaci:
<br>
$$\frac{dx}{dt} = -\frac{1}{\tau_{x}(V)}[x-x_{\infty}(V)],$$
<br>
gdzie:
<br>
$$\tau_{x}(V) = \frac{1}{\alpha_{x}(V) + \beta_{x}(V)},$$
<br>
$$x_{\infty}(V) = \frac{\alpha_{x}(V)}{\alpha_{x}(V) + \beta_{x}(V)}.$$
<br>
Po podstawieniu łatwo zauważyć, że jest to równanie na prawdopodobieństwo otwarcia bramki x.
Poprzez wprowadzenie przybliżeń $b-h \approx a n = w \rightarrow h=w+b \wedge n=\frac{w}{a}$
otrzymujemy równanie:
<br>
$$C\frac{dV}{dt} = -g_{Na}[m_{\infty}(V)]^3(b-w)(V-V_{Na}) - g_{K}\left(\frac{w}{a}\right)^4(V-V_{K})-g_{L}(V-V_{L}) + I,$$
<br>
oraz $\frac{dw}{dt} = -\frac{1}{\tau_{w}(V)}[w-w_{\infty}(V)].$
Mamy więc układ równań:
<br>
$$\frac{dV}{dt} = \frac{1}{\tau}[F(V, w)+RI],$$
<br>
$$\frac{dw}{dt} = - \frac{1}{\tau_{w}(V)}G(V, w),$$
<br>
$R=\frac{1}{g_{L}}$, $\tau = RC$.

### Model Fitzhugh-Nagumo
W tym modelu po przyjęciu odpowiednich postaci funkcji $F$ i $G$ otrzymujemy układ równań:
<br>
$$\frac{dV}{dt} = \tau(V - V^3 - w + I) = f(V, w),$$
<br>
$$\tau \frac{dw}{dt} = V - a - bw = g(V, w).$$
<br>
Takie postacie funkcji są uzasadnione ze względu na to, że bramka m względnie szybko (w porównaniu do innych bramek) wraca do ustalonego stanu, dlatego można przyjąć: $\frac{dm}{dt} = 0$. W sposób eksperymentalny można przyjąć, że $h=h_0=const$.

Na początku szukamy izoklin równań, czyli linii o równaniach $f(x) = c$ oraz $g(x) = d$, gdzie $c=const$ i $d=const$.
Zatem izokliny zerowe ($c=0$ i $d=0$) naszych równań określone są równaniami:
$$V - V^3 - w + I = 0,$$
$$V - a - bw = 0,$$
przy założeniach, że $1 - \frac{2b}{\tau} < a < 1$, $ 0 < b < 1$, $b<\tau^2$.


Z powyższych równań możemy obliczyć wartości $w$:
$$w = V - V^3 + I,$$
$$w = \frac{1}{b}\left(V - a\right).$$
Przecięcia zerowych izoklin są punktami krytycznymi.
Dla dobranych stałych możemy stworzyć wykresy zerowych izoklin odpowiednio dla $I=-0,1$, $I=-0,9$ oraz $I=-1.5$:


```python
import matplotlib.pyplot as plt 
import numpy as np    
```

```python
const1 = [0.7, 0.8, 3, -0.1]
const2 = [0.7, 0.8, 3, -0.9]
const3 = [0.7, 0.8, 3, -1.5]
def isocline(ax, a, b, tau, I, vmin=-1, vmax=1):
    v = np.linspace(vmin,vmax,100)
    ax.plot(v, v - v**3 + I)
    ax.plot(v, (v - a)/b)
```

```python
fig, ax = plt.subplots(1, 1)
isocline(ax, const1[0], const1[1], const1[2], const1[3])
```

```python
fig, ax = plt.subplots(1, 1)
isocline(ax, const2[0], const2[1], const2[2], const2[3])
```

```python
fig, ax = plt.subplots(1, 1)
isocline(ax, const3[0], const3[1], const3[2], const3[3])
```

Możemy również stworzyć pole wektorowe dla układu równań dla odpowiednich wartości $I$:

```python
def model(v, w, var):
    val = np.array([var[2]*(v - v**3 - w + var[3]), (v - var[0] - var[1]*w)/var[2]])
    return val

def vectors(ax, ys, const):
    x = np.linspace(-2, 2, 50)
    y = np.linspace(ys[0], ys[1], 50)
    X, Y = np.meshgrid(x, y)
    dv, dw = model(X, Y, const)
    ax.streamplot(X, Y, dv, dw)

fig, ax = plt.subplots(1, 1)
xs = (-2, 2)
ys = [(1/const1[1])*(x-const1[0])for x in xs]
vectors(ax, ys, const1)
```

```python
fig, ax = plt.subplots(1, 1)
xs = (-2, 2)
ys = [(1/const2[1])*(x-const2[0])for x in xs]
vectors(ax, ys, const2)
```

```python
fig, ax = plt.subplots(1, 1)
xs = (-2, 2)
ys = [(1/const3[1])*(x-const3[0])for x in xs]
vectors(ax, ys, const3)
```

Macież wpółczynnkików dla modelu Fitzhugh-Nagumo wygląda następująco:
$$M = 
\left[\begin{array}{cc}
(1 - V^2)\tau & \tau  \\
\frac{1}{\tau} & \frac{-b}{\tau}  \\
\end{array}\right]
$$
Stabilość zależy od $\lambda$ z równania $|M - \lambda I| = 0$
<br>
$$|M - \lambda I| = \lambda ^2 + [\frac{b}{\tau} - (1 - V^2)\tau]\lambda + [1 - (1 - V^2)b]$$
<br>
Korzystając ze wzorów Viete'a $\lambda_1 \cdot \lambda_2 = [1 - (1 - V^2)b]$, $\lambda_1 + \lambda_{2} = -[\frac{b}{\tau} - (1 - V^2)\tau]$, jeśli $1 - (1 - V^2)b > 0$ oraz $\frac{-b}{\tau} + (1 - V^2)\tau < 0$, to powstaje węzeł stabilny. Natomiast gdy $1 - (1 - V^2)b > 0$ oraz $\frac{-b}{\tau} + (1 - V^2)\tau > 0$ powstaje węzeł niestabilny. Jednak jeśli $1 - (1 - V^2)b < 0$, to tworzy się siodło. Biorąc pod uwagę, że b jest stałe i $\frac{1}{b} > 1$ można zauważyć, że $1 - (1 - V^2)b> 0$ jest zawsze prawdą. Co więcej biorąc pod uwagę założenie $b < \tau^2$ wnioskujemy, że dla $|V| < \sqrt{1 - \frac{b}{\tau^2}}$ punkt jest niestabilny, a dla pozostałych stabilny.
